## Accidents in each state

```python
# plot bar graph for all three years
#sort df by 2021 accidents
df = df.sort_values('Total Number of Persons Injured in Road Accidents on State Highways during 2021', ascending=False)
#drop rows with nan values or 0 values
df = df.dropna()
df = df[df['Total Number of Persons Injured in Road Accidents on State Highways during 2021'] >100]

# remove the total row
df = df[df['States/UTs'] != 'Total']


df.plot.bar(x='States/UTs', y=['Total Number of Persons Injured in Road Accidents on State Highways during 2018',
         'Total Number of Persons Injured in Road Accidents on State Highways during 2019',
         'Total Number of Persons Injured in Road Accidents on State Highways during 2020',
         'Total Number of Persons Injured in Road Accidents on State Highways during 2021'], rot=0, figsize=(20,10))


plt.legend()
plt.xticks(rotation=90)
plt.savefig('plots/project2/bar_state_accidents.png')
plt.show()
```

```python
path = "RTYB1819_A3_4e.csv" # dataset of no of vehicles registered in india of different types
df = pd.read_csv(path)
df.head()
```

```python
df.columns
# Index(['States/Union Territories', 'Two wheelers (I) - Scooters',
#        'Two wheelers (I) - Mopeds', 'Two wheelers (I) - Motor cycles',
#        'Cars (II)', 'Jeeps (III)', 'Omni buses (IV)', 'Tractors (V)',
#        'Trailers (VI)', 'Other vehicles not covered (VII)',
#        'Total Non-Transport (I TO VII)'],
#       dtype='object')
```

```python
# draw a bar plot with percentage of tractor to total non-transport vehicles for each state
df = df.sort_values('Tractors (V)', ascending=False)
df = df.dropna()
df = df[df['Tractors (V)'] >100]

# remove the total row
df = df[df['States/Union Territories'] != 'Total']
df.plot.bar(x='States/Union Territories', y=['Tractors (V)', 'Total Non-Transport (I TO VII)'], rot=0, figsize=(20,10))
plt.legend()
plt.xticks(rotation=90)

plt.savefig('plots/project2/bar_tractor.png')
plt.show()
```

## kmeans clustering 
I tried this to see, if it helps to get any insights

```python
# use k means clustering to find the clusters in states
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

N_CLUSTERS = 4
km = KMeans(n_clusters=N_CLUSTERS)

# drop the total row
df = df[df['States/Union Territories'] != 'Total']
# drop the states column
df_ = df.drop('States/Union Territories', axis=1)
df_.head()
```

```python


# scale the data
scaler = MinMaxScaler()
scaler.fit(df_)
df_scaled = scaler.transform(df_)
df_scaled

# fit the model
km.fit(df_scaled)
km.labels_

# add the labels to the dataframe
df['labels'] = km.labels_
df.head()
```

```python
for i in range(N_CLUSTERS):
    # print names of states in each cluster
    print("Cluster ", i)
    print( df[df['labels']==i] ["States/Union Territories"] )
```

## other experiments 

```python
# create a scatter plot to identify correlation between cars and tractors in each state/union territory
df.plot(kind='scatter', x='Cars (II)', y='Tractors (V)')
plt.title('Correlation between Cars and Tractors in Each State/Union Territory')
plt.xlabel('Number of Cars')
plt.ylabel('Number of Tractors')
plt.grid()
plt.savefig('plots/project2/scatter_cars_tractors.png')
plt.show()
```

```python



# box plot for percentage of each vehicle type to total non-transport vehicles
ratio_df = df.copy()
ratio_df = ratio_df.drop('Total Non-Transport (I TO VII)', axis=1)
ratio_df = ratio_df.drop('labels', axis=1)
ratio_df = ratio_df.drop('States/Union Territories', axis=1)
ratio_df = ratio_df.div(ratio_df.sum(axis=1), axis=0)
ratio_df.plot(kind='box')
plt.title('Distribution of Percentage of Each Vehicle Type to Total Non-Transport Vehicles')
plt.xlabel('Vehicle Type')
plt.ylabel(r'% of Vehicle Type to Total Non-Transport Vehicles')
plt.xticks(rotation=90)
plt.grid()
plt.savefig('plots/project2/boxplot_vehicle_type.png')
plt.show()
```

```python
#make a column with total number of vehicles
df['Total'] = df['Two wheelers (I) - Scooters'] + df['Two wheelers (I) - Mopeds'] + df['Two wheelers (I) - Motor cycles'] + df['Cars (II)'] + df['Jeeps (III)'] + df['Omni buses (IV)'] + df['Tractors (V)'] + df['Trailers (VI)'] + df['Other vehicles not covered (VII)']
#ratio of tractors to total vehicles
df['tractor_percentage'] = df['Tractors (V)']/df['Total']

#make box plot of tractor percentage 
df.plot(kind='box', y=['tractor_percentage'])
plt.title('Distribution of Percentage of Tractors across Different States/Union Territories')
plt.xlabel('Percentage of Tractors')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=90)
plt.savefig('plots/project2/boxplot_tractor_percentage.png')
plt.show()


```

```python
# make bar plot with different vehicle types in different colors for each state/union territory and sort by total number of vehicles


df = df.sort_values('Total', ascending=False)
df.plot(kind='bar', x='States/Union Territories', y=['Two wheelers (I) - Scooters', 
                                                      'Two wheelers (I) - Mopeds', 
                                                      'Two wheelers (I) - Motor cycles',
                                                      'Cars (II)', 
                                                      'Jeeps (III)', 
                                                      'Omni buses (IV)', 
                                                      'Tractors (V)',
                                                      'Trailers (VI)', 
                                                      'Other vehicles not covered (VII)'], stacked=True)
plt.title('Number of Vehicles in Different States/Union Territories')
plt.savefig('plots/project2/segmented_barplot_vehicle_types.png')
```

```python
# cluster plot of tractor percentage and two wheeler percentage
df.plot(kind='scatter', x='Tractors (V)', y='Two wheelers (I) - Motor cycles')
#make color map of each state
color_map = {0:'red', 1:'blue', 2:'green', 3:'yellow'}
#make scatter plot of each state with different color
for i in range(N_CLUSTERS):
    plt.scatter(df[df['labels']==i]['Tractors (V)'], df[df['labels']==i]['Two wheelers (I) - Motor cycles'], color=color_map[i])
#make legend
plt.legend()


plt.title('Correlation between Tractors and Two Wheelers')
plt.xlabel('Number of Tractors')
plt.ylabel('Number of Two Wheelers')
plt.grid()
plt.savefig('plots/project2/scatter_tractor_two_wheeler.png')
plt.show()

```
